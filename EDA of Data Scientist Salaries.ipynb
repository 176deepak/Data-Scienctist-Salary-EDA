###### replacing column's value with their full form or other values  

###### 1.For experience_level column  

```python
df["experience_level"] = df["experience_level"].replace({"SE":"Senior Engineer", "MI" : "Intermidiate", "EN":"Entry level", "EX":"Experienced"})
```

###### 2.For employment_type column 

```python
df["employment_type"] = df["employment_type"].replace({"FT":"Full time", "PT":"Part time", "FL":"Freelance", "CT":"Contract based"})
```

###### 3.For remote_ratio column 

```python
df["remote_ratio"] = df["remote_ratio"].astype(str)
```

```python
df["remote_ratio"] = df["remote_ratio"].replace({"0":"On-site", "50":"Half-Remote", "100":"Full-Remote"})
```

###### 4.For company_size column 

```python
df["company_size"] = df["company_size"].replace({"S":"Small", "M":"Medium", "L":"Large"})
```

###### function replacing country codes values of employee_residence and company_location with their corresponding names 

```python
def country_name(country_code):
    try:
        country = cc.countries.get(alpha_2=country_code)
        return country.name
    except LookupError:
        return "Unknown"
```

```python
df['employee_residence'] = df['employee_residence'].apply(country_name)
```

```python
df['company_location'] = df['company_location'].apply(country_name)
```

###### function for assigning job_title column values to different different groups 

```python
def job_groups(job):
    data_engineering = ["Data Engineer", "Data Analyst", "Analytics Engineer", "BI Data Analyst", "Business Data Analyst", 
                        "BI Developer", "BI Analyst", "Business Intelligence Engineer", "BI Data Engineer", "Power BI Developer"]
    data_scientist = ["Data Scientist", "Applied Scientist", "Research Scientist", "3D Computer Vision Researcher", 
                      "Deep Learning Researcher", "AI/Computer Vision Engineer", "Principal Data Scientist"]
    machine_learning = ["Machine Learning Engineer", "ML Engineer", "Lead Machine Learning Engineer", 
                        "Principal Machine Learning Engineer"]
    data_architecture = ["Data Architect", "Big Data Architect", "Cloud Data Architect", "Principal Data Architect"]
    management = ["Data Science Manager", "Director of Data Science", "Head of Data Science", "Data Scientist Lead", 
                  "Head of Machine Learning", "Manager Data Management", "Data Analytics Manager"]

    if job in data_engineering:
        return "Data Engineering"
    elif job in data_scientist:
        return "Data Science"
    elif job in machine_learning:
        return "Machine Learning"
    elif job in data_architecture:
        return "Data Architecture"
    elif job in management:
        return "Management"
    else:
        return "Other"
```

```python
df['job_class'] = df['job_title'].apply(job_groups)
```

# *Data Visualization* 

#### *Univariate Analysis* 

```python
#To plot the distribution of each class of individual columns in the dataset.

def percentage_plot(data, title, y_label, color):
    fig, ax = plt.subplots(figsize=(14, 6))
    ax.barh(data.index[:6], data.values[:6], color=color)
    for i in ax.patches:
        plt.text(i.get_width()+0.2, i.get_y()+0.35,
                 str(round((i.get_width()), 2)),
                 fontsize=10, fontweight='bold',
                 color='red')
    ax.set_xlabel('Percentage', fontsize=13)
    ax.set_ylabel(y_label, fontsize=13)
    ax.set_title(title, fontsize=15)
    plt.show()
```

###### barchart for job_class column 

```python
job_class_percent = df['job_class'].value_counts(normalize=True)*100
percentage_plot(data=job_class_percent, title='Job Category %', y_label='Job Category', color=['bisque', 'darkorange', 'burlywood', 'darkgoldenrod', 'khaki'])
```

result: in dataset more number of jobs belong to Data Engineering Job and then Data Science Job

Data Engineering : 44.16 & Data Science : 25.04
